```python
import numpy as np
import pandas as pd
```

```python
data_path= r'../input/brasilian-houses-to-rent/houses_to_rent.csv'
data= pd.read_csv(data_path)
```

```python
data.head(5)
```

* Things to do
* remove 'Unnamed: 0' column
* convert all the categorical values(string values)'animal','furniture' to numbers
* remove all the R$ on the columns lie 'hoa','rent amount', property tax','fire insurance', 'total'
* change all '-' to ''

```python
#remove/drop 'Unnamed: 0' column
print("The list of columns:> ",data.columns)
data.drop(data.columns[0],axis=1, inplace=True)
data.head(5)
```

```python
#convert all the categorical values(string values)'animal','furniture' to numbers
data['animal'].replace(to_replace='acept', value=1,inplace=True)
data['animal'].replace(to_replace='not acept', value=0,inplace=True)

data['furniture'].replace(to_replace='furnished', value=1,inplace=True)
data['furniture'].replace(to_replace='not furnished', value=0,inplace=True)
```

```python
data.head()
```

```python
#remove all the R$ on the columns lie 'hoa','rent amount', property tax','fire insurance', 'total'
columns= ['hoa','rent amount','property tax','fire insurance','total']
for column in columns:
    data[column].replace(to_replace='R\$', value='',regex=True,inplace=True) #$ should be escaped 
    data[column].replace(to_replace=',', value='',regex=True,inplace=True)
```

```python
data.head(5)
```

```python
#change all '-' to ''
data['floor'].replace(to_replace='-',value=0,inplace=True)
```

```python
data.head(5)
```

```python
# in the column 'hoa' there are more categorical values like 'Sem info', 'Incluso' so need to replace them
data['hoa'].value_counts()
```

```python
data['hoa'].replace(to_replace='Sem info',value=0,inplace=True)
data['hoa'].replace(to_replace='Incluso',value=0,inplace=True)
data['property tax'].replace(to_replace='Incluso', value='0', inplace=True)
```

```python
data.head(5)
```

```python
#converting all the data to int64 dtype
data= data.astype(dtype=np.int64)
data.head(5)
```

# Search functionality on the data

```python
data.isin(['Incluso']).any() #search for 'Incluso'
```

# Shuffle the data

```python
data= data.sample(frac=1).reset_index(drop=True)
data.head(5)
```

# Remove Y from the data

```python
y= data['city']
x= data.drop('city',axis=1)
```

```python
print("Shape of Y:> ",y.shape)
print("Shape of X:> ",x.shape)
```

# Scaling the dataset to 0-1

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler= MinMaxScaler()
scaler.fit(x)
x= scaler.transform(x)
```

# Splitting the dataset

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test= train_test_split(x,y,train_size=0.8)
```

# Training the AI models

```python
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.neural_network import MLPClassifier
```

```python
# initialising the model
logistic_model= LogisticRegression(penalty='l2', verbose=1)
support_vector_model= SVC(kernel='rbf', verbose=1)
perceptron_model= MLPClassifier(hidden_layer_sizes=(16, 16), activation='relu', solver='adam', verbose=1)
```

```python
#fitting the model on the dataset
logistic_model.fit(x_train,y_train)
support_vector_model.fit(x_train,y_train)
perceptron_model.fit(x_train,y_train)
```

# Scoring the model on the testing dataset

```python
print(logistic_model.score(x_test,y_test))
print(support_vector_model.score(x_test,y_test))
print(perceptron_model.score(x_test,y_test))
```
